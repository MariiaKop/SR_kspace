````python
def npy_load(path):
    with open(path, 'rb') as f:
        return np.load(f)
    

RANDOM_STATE = 42
````

````python
class DataLoader(torch_data.Dataset):
    def __init__(self, path_to_data, seed=None):
        super().__init__()

        self.images = [os.path.join(path_to_data, file) for file in os.listdir(path_to_data) 
                       if file.endswith('.npy')]
    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        
        img = npy_load(self.images[idx])
        return self.images[idx], img
````
